```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew


# Load your dataset
df = pd.read_csv("D:\\collegePlace.csv")

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
```

```python
## 1. Pie Chart - Placement Distribution
plt.figure(figsize=(8, 8))
placement_counts = df['PlacedOrNot'].value_counts()
plt.pie(placement_counts, labels=['Placed', 'Not Placed'], autopct='%1.1f%%', 
        colors=['#66b3ff','#ff9999'], startangle=90)
plt.title('Placement Distribution', fontsize=16)
plt.show()
```

```python
# 2. Box Plots - Key Factors vs Placement
plt.figure(figsize=(15, 10))

# CGPA
plt.subplot(2, 2, 1)
sns.boxplot(x='PlacedOrNot', y='CGPA', data=df)
plt.title('CGPA Distribution by Placement Status')
plt.xticks([0, 1], ['Not Placed', 'Placed'])


plt.tight_layout()
plt.show()
```

```python
## 3. Bar Charts with Skewness
# Calculate skewness for numerical features
numerical_features = ['Age', 'CGPA', 'Internships', 'HistoryOfBacklogs']
skewness = df[numerical_features].apply(lambda x: skew(x.dropna()))



# Bar chart for Stream vs Placement
plt.figure(figsize=(12, 6))
stream_placement = pd.crosstab(df['Stream'], df['PlacedOrNot'])
stream_placement.plot(kind='bar', stacked=True, color=['#ff9999','#66b3ff'])
plt.title('Placement by Stream', fontsize=16)
plt.xlabel('Stream', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)
plt.legend(['Not Placed', 'Placed'])
plt.show()

```

```python
## 4. Scatter Plots - Relationships between key factors
plt.figure(figsize=(15, 10))



# Internships vs HistoryOfBacklogs
plt.subplot(2, 2, 3)
sns.scatterplot(x='Internships', y='HistoryOfBacklogs', hue='PlacedOrNot', data=df, palette=['#ff9999','#66b3ff'])
plt.title('Internships vs Backlog History by Placement Status')



plt.tight_layout()
plt.show()
```

```python
## 5. Heatmap - Correlation between features
plt.figure(figsize=(10, 8))
# Select only numerical columns for correlation
numerical_df = df.select_dtypes(include=['number'])
correlation = numerical_df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Features', fontsize=16)
plt.show()

# Additional Analysis: Hostel vs Placement
plt.figure(figsize=(8, 6))
hostel_placement = pd.crosstab(df['Hostel'], df['PlacedOrNot'])
hostel_placement.plot(kind='bar', stacked=True, color=['#ff9999','#66b3ff'])
plt.title('Placement by Hostel Stay', fontsize=16)
plt.xlabel('Hostel Stay (1=Yes, 0=No)', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(['Not Placed', 'Placed'])
plt.show()
```
