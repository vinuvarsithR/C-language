```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("D:\\collegePlace.csv")  # update path if needed
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x="Gender", y="PlacedOrNot", estimator=lambda x: sum(x)/len(x))
plt.title("Placement Rate by Gender")
plt.ylabel("Placement Rate")
plt.ylim(0, 1)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Stream", y="PlacedOrNot", estimator=lambda x: sum(x)/len(x))
plt.title("Placement Rate by Academic Stream")
plt.ylabel("Placement Rate")
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x="Internships", y="PlacedOrNot", estimator=lambda x: sum(x)/len(x))
plt.title("Placement Rate vs. Number of Internships")
plt.xlabel("Number of Internships")
plt.ylabel("Placement Rate")
plt.ylim(0, 1)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x="PlacedOrNot", y="CGPA")
plt.title("CGPA Distribution vs. Placement Status")
plt.xlabel("Placed (1) vs Not Placed (0)")
plt.ylabel("CGPA")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x="HistoryOfBacklogs", y="PlacedOrNot", estimator=lambda x: sum(x)/len(x))
plt.title("Placement Rate vs. History of Backlogs")
plt.xlabel("History of Backlogs (1 = Yes, 0 = No)")
plt.ylabel("Placement Rate")
plt.ylim(0, 1)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
